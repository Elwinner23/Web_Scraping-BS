```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://bina.az/'
response = requests.get(url)
soup =  BeautifulSoup(response.text, 'html.parser')
```

```python
links, prices, places, date = [], [], [], []
garb = soup.findAll('a', attrs = {"class": "item_link"})
domain = 'https://bina.az'
for garb_names in garb:
    link = domain+garb_names.get('href')
    r2 = requests.get(link)
    c = r2.content
    soup2 = BeautifulSoup(c, "html.parser")
    
    prices_cont = soup2.find('p', attrs = {"class": "azn"}).getText(strip = True)
    places_cont = soup2.find('div', attrs = {"class" : "map_address"}).getText()
    date_cont = soup2.find('div', attrs = {"class" : "item_info"}).getText()
    links.append(link)
    prices.append(prices_cont)
    places.append(places_cont)
    date.append(date_cont)
```
